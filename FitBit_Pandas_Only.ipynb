```python
#3. convert all the dates avaible in dataset to timestamp format in pandas and in sql you to convert it in date format
data['converted_activity_date'] = pd.to_datetime(data['ActivityDate'])
```

```python
#4 . Find out in this data that how many unique id's we have 
len(data.Id.unique())
```

```python
# 5. which id is one of the active id that you have in whole dataset
a = pd.DataFrame(data['Id'].value_counts())
a = a.reset_index()
for i in range(len(a['Id'])):
    if (max(a['Id'])==(int(a[i:i+1]['Id']))):
        print("the most active person's ID is : ",int(a[i:i+1]['index']))
```

```python
#6 . how many of them have not logged there activity find out in terms of number of ids 
len(set(data["Id"])) - len(set(data[data['LoggedActivitiesDistance']>0].Id.to_list()))
```

```python
#7. which id is one of the laziest id that you have in whole dataset
a = pd.DataFrame(data['Id'].value_counts())
a = a.reset_index()
for i in range(len(a['Id'])):
    if (min(a['Id'])==(int(a[i:i+1]['Id']))):
        print("the most laziest person's ID is : ",int(a[i:i+1]['index']))
```

```python
#8 . Explore over an internet that how much calories burn is required for a healthy person and find out how many healthy person we have in our dataset 
len(set(data[data['Calories']>2500].Id.to_list()))
```

```python
#9. how many person are not a regular person with respect to activity try to find out those 
#assuming a normal person would take 8000 steps per day and dataset is of 21 days so total steps per person should be 21*8000=1,68,000
q = pd.DataFrame(data.groupby('Id')['TotalSteps'].sum())
q.reset_index()
not_normal_count = len(q[q['TotalSteps'] < 168000].index.to_list())
print(not_normal_count,  " this many no of peoples are not a regular person with respective to the Total Steps")
```

```python
#10 . who is the thired most active person in this dataset find out those in pandas and in sql both . 
a = pd.DataFrame(data['Id'].value_counts())
a = a.reset_index()
third_most = 15
for i in range(len(a['Id'])):
    if int(a[i:i+1]['Id'])==third_most:
        print("third most active person's ID is : ",int(a[i:i+1]['index']))
```

```python
#11 . who is the 5th most laziest person avilable in dataset find it out 
a = pd.DataFrame(data['Id'].value_counts())
a = a.reset_index()
fifth_most = 12
for i in range(len(a['Id'])):
    if (fifth_most==(int(a[i:i+1]['Id']))):
        print("fifth most laziest person's ID is : ",int(a[i:i+1]['index']))
```
