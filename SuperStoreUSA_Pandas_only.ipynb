```python
#4 . try  to join order and return data both in sql and pandas
orders_returns = orders.join(returns, lsuffix="_Order", rsuffix='_return')
orders_returns
```

```python
#5 . Try to find out how many unique customer that we have 
unique_cust_count = len(set(orders['Customer Name'].to_list()))
print("total no of unique customers count is : ", unique_cust_count)
```

```python
ship_mode = set(orders['Ship Mode'].to_list())
ship_mode = list(ship_mode)
print("There are ",len(ship_mode), " different modes of shipment available")
```

```python
#7 . find out how many different differnet shipement mode that we have and what is a percentage usablity of all the shipment mode with respect to dataset 
from collections import Counter
status_d=Counter(orders['Ship Mode'].to_list())
print(status_d)

key_list = list(status_d.keys())
val_list = list(status_d.values())

fig1, ax1 = plt.subplots()
ax1.pie(val_list, labels=key_list, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
#8 . Create a new coulmn and try to find our a diffrence between order date and shipment date

orders['difference_ship_order_date'] = orders['Ship Date'] - orders['Order Date']
```

```python
#9 . base on question number 8 find out for which order id we have shipment duration more than 10 days 
# As we have this 'difference_ship_order_date' column in timedelta datatype so first we need to convert it into int 
orders['difference_ship_order_date'] = orders['difference_ship_order_date'].dt.days.astype('int16')
#after converting this we can give the condition i.e. > 10
orders[orders['difference_ship_order_date'] > 10]['Order ID'].to_list()
```

```python
#11 . Gorup by region and find out which region is more profitable 
region_profit = pd.DataFrame(orders.groupby('Region')['Profit'].sum())
region_profit.reset_index(inplace = True)
most_profitable_region = region_profit[region_profit['Profit']==max(region_profit['Profit'])].Region.to_list()
print('The most profitable region is ', most_profitable_region[0]," region")
```

```python
#12 . Try to find out overalll in which country we are giving more didscount
# Here we don't have country so we fill extract for cities
state = pd.DataFrame(orders.groupby('State or Province')['Discount'].sum())
state.reset_index(inplace=True)
most_discounted_state = state[state['Discount']==max(state['Discount'])]['State or Province'].to_list()
print("The most discount offered in ",most_discounted_state[0])
```

```python
#13 . Give me a list of unique postal code 
orders['Postal Code'].unique()
```

```python
# 14 . which customer segement is more profitalble find it out . 
segment = pd.DataFrame(orders.groupby('Customer Segment')['Profit'].sum())
segment.reset_index(inplace=True)
profitable_segment = segment[segment['Profit']==max(segment['Profit'])]['Customer Segment'].to_list()
print("The most profitable customer segment is ",profitable_segment[0])
```

```python
# 15 . try to find out the 10th most loss making product catagory .
prod_cat = pd.DataFrame(orders.groupby('Product Sub-Category')['Profit'].sum())
prod_cat.reset_index(inplace=True)
loss_prod_cad = prod_cat[prod_cat['Profit']==min([prod_cat['Profit']])]['Product Sub-Category'].to_list()
print('The product category in loss is ',loss_prod_cad[0]," with loss to total ",min(prod_cat['Profit']))
```

```python
##16 . Try to find out 10 top  product with highest margins 
orders['Product Base Margin'].fillna(method = 'Backfill',inplace = True) 
prod_margin1 = pd.DataFrame(orders.groupby('Product Name')['Product Base Margin'].sum())
prod_margin1.reset_index(inplace=True)
prod_margin1.sort_values('Product Base Margin',inplace = True)
top_10 = prod_margin1[-10:]
top_10.sort_values('Product Base Margin',inplace = True, ascending=False)
top_10['Product Name'].to_list()
```
